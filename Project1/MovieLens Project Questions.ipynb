### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use `sum(axis=1)` store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.  

```python
# Visudalize how genre populatroty has changed over the years.
# USe genre as a percentage of releases per year to determine this.

g_item = item.copy()   # Copy item into a temporary data frame as do not wna to destroy orginal DF. 

# Take the DD-Mon-YYYY format and convert it to individual columms.
# Place these colmns ina new table so as to not destoy orginal data
g_item['release year'] = g_item['release date'].str.split('-').str[2]   #Split out the year 



g_item.drop(columns=['movie id', 'movie title', 'release date'], inplace=True) # Drop unused columns
g_item.set_index(g_item['release year'], inplace=True) #S et the index otthe relase year so we can group these by year

g_item = g_item.groupby(g_item['release year']).sum()  #Group by year and then sum remain columns to get totla movies per genre per year
g_item['total movies'] = g_item.sum(axis=1) # Sum all movies ot get a tioal count so can figure percentage.
g_item = g_item.div(g_item['total movies'], axis= 0) * 100   #Determine percentage and display as %
print(g_item)  #Print the table just for sanity sake
```

```python
# Display a heat map to vusally observe how the mvoie genre popularity has changed

# Protecting code via teh if statement as notice multiple runs can casue erros as colmn had been previously removed.
# POtentially fix this by setting up datamap in same cell but want the previous graph display as well and seems eaier unde  multiple cells.
if 'total movies' in g_item.columns:
    g_item.drop(columns = "total movies", inplace=True) # Dop the "total movies column" as doesn't really porvide any data ot the graph
       
f, ax = plt.subplots(figsize=(30, 25))
plt.xticks(rotation=90)           # Rotate year dispaly for clairity

print(sns.heatmap(g_item, annot=True, linewidths=.5, ax=ax).set_title("Number of movies per genre by year"))
```

```python
# Find top 25 movies with at least 100 ratings

# Assignign a temp data rame to data as to not destroy orginal dataframe which wil be needed in next question.
# This section is to get the total number of ratings per movie. It will be used to get items with more than 100 movies later
# tmppdata propbably isn't needed and could be included into the a_data df, whcih woudl ermove the merge but this was "safer" and easier
# to implement.
tmpdata = data.copy()
tmpdata['Rating Count'] = 1   # Add a column named arting count to allow average calculation later. Will be based on total movies fopr that row
tmpdata.drop(columns=['user id', 'rating','timestamp'], inplace=True) #Drop uneeded columns
tmpdata = tmpdata.groupby('movie id').sum()  #
tmpdata.reset_index(inplace=True)

# Assigning a temp dataframe to data as to not destroy orginal dataframe which wil be needed in next question.
a_data = data.groupby(data['movie id']).mean()   #get the average rating per movie
a_data.rename(columns={"rating": "Average Rating"}, inplace=True) #Rename the column just for accuracy.
a_data.drop(columns=['user id','timestamp'], inplace=True) # Delete uneeded columns

a_data.reset_index(inplace=True) # Setting up to allow easier merging

a_data= a_data.merge(tmpdata, left_on="movie id",right_on="movie id")  # Merge the average data frame with the total count data frame


nogo = a_data[ a_data['Rating Count'] <= 100 ].index  # Determine the index of all movies in dataframe that do not have more than 100 ratings
a_data.drop(nogo , inplace=True) # Delete these row indexes from dataFrame

# Need to merge  item dataframe in to get movie titles for display.
a_data= a_data.merge(item, left_on="movie id",right_on="movie id")
a_data.drop(columns=['movie id', 'release date','unknown','Action','Adventure','Animation','Childrens','Comedy','Crime','Documentary','Drama','Fantasy','Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western'], inplace=True) # Drop unuised columns

a_data.set_index('movie title', inplace=True) # Set movie tiotle as index justto make printign look better.
a_data.sort_values(by=['Average Rating'],ascending=False).head(25) # Display Top 25 Average rated movies
```

```python
# Check gender distribution for three categories of movies.
#First set up a singular dataframe of all three (user, item and data) other datframes
#Common keys are:
#    user <-> data = user_id
#    item <-> (user <-> data) = movie_id

# validating that each merge is 1 to many ot ensure that comon dat gets distributed to each row that has a same key.
newdf = user.merge(data, left_on='user id', right_on='user id', validate="1:m")
newdf = item.merge(newdf, left_on='movie id', right_on='movie id', validate="1:m")

# For the next several calculations, set a new datframe for each category based on "newdf" rather than use "newdf" itslef so can maintain its integrity for future calcs.
```

```python
########### DRAMA
# Check Male to Female RAW numbers for Drama movies

newdf1 = newdf.groupby(["gender","Drama"])["Drama"].count()
newdf1
```

```python
# Determine percentage of men and women in Drama
# Note: This is GENDER / (GENDER RATE + GENDER NO RATE) therefore it is percentage of each gender and ot of total number of people
# In other words although there are fewer women making ratings the data is normalized with men so percentage have same weigth.
wy = (newdf1.iloc[1] / (newdf1.iloc[0] + newdf1.iloc[1]))*100  # F Rows (0 = no rating, 1 = rating)
my = (newdf1.iloc[3] / (newdf1.iloc[2] + newdf1.iloc[3]))*100  # M Rows (0 = no rating, 1 = rating)

print("Male Percentage = %s, Female Percentage = %s. Men watch more Drama is %s" % (my, wy, my > wy))
```
